```python
"""
Characteristics of this dataset:

This fossil fuel data contains co2 emissions by country and year. It contains the total fuel, solid fuel, liquid fuel, gas fuel, cement, gas flaring, and bunker fuels for each country.
Overall, there are 17233 rows in the dataset.
Some of the data is string, and some of them are numbers.
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv',encoding='iso-8859-1')
df[:100]
```

```python
# Chart 1: Bar chart of highest per capita for top 10 countries over all of the years

per_capita_by_country = df.groupby('Country')["Per Capita"].sum()
sorted_values = per_capita_by_country.sort_values(ascending=False)
top_10=sorted_values.head(10)
top_10.plot(title= "Highest per capita (10 countries)", kind="bar", color = 'blue')
plt.xlabel('Country')
plt.ylabel('Per Capita')
plt.show()
```

```python
# Chart 2: Bar chart of highest liquid fuel for top 10 countries over all of the years

liquid_fuel_by_country = df.groupby('Country')["Liquid Fuel"].sum()
sorted_values = liquid_fuel_by_country.sort_values(ascending=False)
top_10=sorted_values.head(10)
top_10.plot(title= "Highest Liquid Fuel (10 countries)", kind="bar", color = 'orange')
plt.ylim(0, top_10.max()*1.2)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.xlabel('Country')
plt.ylabel('Liquid Fuel')
plt.show()
```

```python
# Chart 3: Bar chart of highest solid fuel for top 10 countries over all of the years

solid_fuel_by_country = df.groupby('Country')["Solid Fuel"].sum()
sorted_values = solid_fuel_by_country.sort_values(ascending=False)
top_10=sorted_values.head(10)
top_10.plot(title= "Highest Solid Fuel (10 countries)", kind="bar", color = 'pink')
plt.ylim(0, top_10.max()*1.2)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.xlabel('Country')
plt.ylabel('Solid Fuel')
plt.show()
```
