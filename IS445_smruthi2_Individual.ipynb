```python
"""
Characteristics of this dataset:

This fossil fuel data contains co2 emissions by country and year. It contains the total fuel, solid fuel, liquid fuel, gas fuel, cement, gas flaring, and bunker fuels for each country.
Overall, there are 17233 rows in the dataset.
Some of the data is string, and some of them are numbers.
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
```

```python
df = pd.read_csv('dataset.csv',encoding='iso-8859-1')
df[:100]
```

```python
### This tree map shows the top 15 countries with the highest gas fuel.
plt.figure(figsize=(20,8))
gas_fuel_by_country = df.groupby('Country')["Gas Fuel"].sum()
gas_fuel_by_country_sorted = gas_fuel_by_country.sort_values(ascending = False)
top_15 = gas_fuel_by_country_sorted.head(15)
squarify.plot(sizes = top_15.values, label=top_15.index, alpha = 0.5)
plt.title("Top 15 countries with highest gas fuel")
```

```python
### This tree map shows the top 15 countries with the highest gas, liquid and solid fuel.
plt.figure(figsize=(25,8))
fuel_by_country = df.groupby('Country')[["Solid Fuel", "Liquid Fuel", "Gas Fuel"]].sum()
fuel_by_country_sorted =fuel_by_country.sum(axis=1).sort_values(ascending = False)
top_15 = fuel_by_country_sorted.head(15)
squarify.plot(sizes = top_15.values, label=top_15.index, alpha = 0.5)
plt.title("Top 15 countries with highest solid/liquid/gas fuel")
```

```python
### The pie chart that shows the top 10 countries based on the highest solid fuel

plt.figure(figsize=(25,8))
fuel_by_country = df.groupby('Country')["Solid Fuel"].sum()
top_10 = fuel_by_country_sorted.head(10)
plt.pie(top_10.values, labels=top_10.index, shadow = True)
plt.title("Top 10 countries with highest solid fuel")

```
